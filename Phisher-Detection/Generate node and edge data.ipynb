```python
import pandas as pd
import numpy as np
```

```python
df_file = "/content/drive/My Drive/BaselineToShow/D1.csv"
df = pd.read_csv(df_file)
```

```python
df.head()
```

```python
df_node = pd.DataFrame(columns = ['node', 'isp'])
```

```python
unique_from = df['from'].unique()
unique_from
```

```python
for i in range(0, unique_from.shape[0]):
  is_p = df.loc[df['from'] == unique_from[i]]['fromIsPhi'].unique()[0]
  df_add = {'node': unique_from[i], 'isp': is_p}
  df_node = df_node.append(df_add, ignore_index = True)
```

```python
df_node
```

```python
unique_to = df['to'].unique()
unique_to
```

```python
df_node
df_node.shape
```

```python
for i in range(0, unique_to.shape[0]):
  if (unique_to[i] not in unique_from):
    is_p = df.loc[df['to'] == unique_to[i]]['toIsPhi'].unique()[0]
    df_add = {'node': unique_to[i], 'isp': is_p}
    df_node = df_node.append(df_add, ignore_index = True)
```

```python
df_node
df_node.shape
```

```python
df_node.to_csv("/content/drive/My Drive/BaselineToShow/nodeD1.csv", encoding='utf-8', index=False)
```
