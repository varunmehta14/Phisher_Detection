```python
def Diff(li1, li2):
    li_dif = [i for i in li1 + li2 if i not in li1 or i not in li2]
    return li_dif
```

```python
fin_nodes = []
fin_nodes.append(sel_node[0])
len_list = 1
need = 1000
i = 0

while need > len_list:
  # print(i)
  # list_request = []
  if i > fin_nodes.__len__()-1:
    all_nodes = Diff(all_nodes, fin_nodes)
    beta = random.choice(all_nodes)
    print('diff lena pada')
  else:
    beta = fin_nodes[i]
  list_request = [n for n in nx_g[beta]]
  list_req = []
  # print(len(list_req))
  for node in list_request:
    if node not in fin_nodes:
      list_req.append(node)
  # print('p')
  # print(type(list_req))
  size = len(list_req)
  # print(list_req, size)
  if (len_list + size > need):
    req = need - len_list
    fin_nodes = fin_nodes + list_req[0:req]
    len_list += req
  else:
    fin_nodes = fin_nodes + list_req
    len_list += size
  # print("fin_nodes", len_list, fin_nodes)
  print(len_list)
  i+=1
```

```python
len(fin_nodes)
sel_nodes = fin_nodes
```

```python
sel_node_g = nx_g.subgraph(sel_nodes)
nx.info(sel_node_g)
```

```python
new_nodes = list(sel_nodes)
```

FT0: IsP = If that node is a Phisher Node \
FT1: In-Degree = The number of in-going edges \
FT2: Out-Degree = The number of out-going edges \
FT3: Degree = The total number of edges

```python
isp = []
indegree = []
outdegree = []
degree = []
for i in sel_nodes:
  df_new = features.loc[features['node'] == i]
  isp.append(df_new['isp'].values[0])
  indegree.append(df_new['in_degree'].values[0])
  outdegree.append(df_new['out_degree'].values[0])
  degree.append(df_new['degree'].values[0])
```

```python
type(isp)
isp
```

FT4: In-Strength = The total amount of ETH sent to the current node

```python
instrength = []
for i in sel_nodes:
  print(i)
  df_new = df.loc[df['to'] == i]
  total = df_new['amount'].sum()
  instrength.append(total)
```

FT5: Out-Strength = The total amount of current nodes transfer to other nodes

```python
outstrength = []
for i in sel_nodes:
  df_new = df.loc[df['from'] == i]
  total = df_new['amount'].sum()
  outstrength.append(total)
```

FT6: Strength = The sum of edgesâ€™ weights to the node, aka the nodeâ€™s total transaction amount

```python
strength = []
for i in range(0, len(instrength)):
  strength.append(instrength[i] + outstrength[i])
```

FT7: Number of neighbors

```python
numberneightbours = []
for i in sel_nodes:
  neighbours = [n for n in sel_node_g[i]]
  numberneightbours.append(len(neighbours))
```

FT8: The Inverse of the Transaction Frequency = It is defined as the time interval between the accountâ€™s first and last transaction divided by FT3

```python
df
```

```python
# sel_nodes
node=0
df.loc[df['from'] == node]
```

```python
invtransfreq = []
for node in sel_nodes:
  node_df_1 = df.loc[df['from'] == node]
  node_df_2 = df.loc[df['to'] == node]
  frames = [node_df_1, node_df_2]
  node_df = pd.concat(frames)
  min = node_df['timestamp'].min()
  max = node_df['timestamp'].max()
  invtransfreq.append(max-min)

for i in range(0,len(invtransfreq)):
  invtransfreq[i] = invtransfreq[i]/degree[i]
```

```python
data = {
    "node": new_nodes,
    "isp": isp,
    "indegree": indegree,
    "outdegree": outdegree,
    "degree": degree,
    "instrength": instrength,
    "outstrength": outstrength,
    "strength": strength,
    "numberneightbours": numberneightbours,
    "invtransfreq": invtransfreq
}
df_10k = pd.DataFrame(data)
```
